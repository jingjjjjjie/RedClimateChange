```python
import pandas as pd
import os
from ipywidgets import widgets, HBox, VBox
from IPython.display import display, clear_output
from PIL import Image
import matplotlib.pyplot as plt
```

```python
postsdf = pd.read_csv('./../metadata/Red_image posts full meta.csv')
```

```python
def imagereturn(id, relpath):
    imagelist = []
    directory_path = os.path.join(relpath, id) 
    
    try:
        items = os.listdir(directory_path)
    except FileNotFoundError:
        return "Directory not found. Please check the path."
    
    for i, item in enumerate(items):
        if item.endswith('.jpg'):  
            new_path = os.path.join(directory_path, item) 
            imagelist.append(new_path)
    
    return imagelist

def get_keywords_by_id(df, image_id):
    result = df[df['note_id'] == image_id]
    if not result.empty:
        return result['keywords'].values[0]  
    return None

def get_title_by_id(df, image_id):
    result = df[df['note_id'] == image_id]
    if not result.empty:
        return result['og:title'].values[0]  
    return None

def get_description_by_id(df, image_id):
    result = df[df['note_id'] == image_id]
    if not result.empty:
        return result['description'].values[0] 
    return None
```

```python
from ipywidgets import widgets, HBox, VBox
from IPython.display import display, clear_output
from PIL import Image
import matplotlib.pyplot as plt

# Function to display an image without white background or axis
def show_image(image_path):
    img = Image.open(image_path)
    
    # Create a Matplotlib figure
    fig, ax = plt.subplots(figsize=(5, 5))  # Adjust figure size as needed
    ax.imshow(img)
    ax.axis('off')  # Turn off axes to remove white background and axis labels
    fig.patch.set_visible(False)  # Remove figure background
    ax.margins(0)  # Remove margins
    plt.gca().set_position([0, 0, 1, 1])  # Full-space image display
    plt.show()

# Create the image carousel with navigation buttons, title, description, and keywords
def image_carousel(image_paths, titles, descriptions, keywords_list):
    current_index = 0
    
    # Create next and previous buttons with professional style
    next_button = widgets.Button(description="Next", button_style='success')
    prev_button = widgets.Button(description="Previous", button_style='success', disabled=True)
    
    # Output widget to display images
    output = widgets.Output()
    
    # Title, description, and keywords widgets (using HTML for clean and styled text)
    title_box = widgets.HTML(value="<b>Title:</b> " + titles[current_index], 
                             layout=widgets.Layout(width='400px', height='auto'))
    
    # Updated: Adjust description box height to allow full display
    description_box = widgets.HTML(value="<b>Description:</b> " + descriptions[current_index], 
                                   layout=widgets.Layout(width='400px', height='200px', overflow='auto'))  # Fixed height and scroll
    
    keywords_box = widgets.HTML(value="<b>Keywords:</b> " + keywords_list[current_index], 
                                layout=widgets.Layout(width='400px', height='auto'))
    
    # Function to update the displayed image, buttons, and information
    def update_image(change=None):
        nonlocal current_index
        
        with output:
            clear_output(wait=True)
            show_image(image_paths[current_index])
        
        # Update title, description, and keywords
        title_box.value = "<b>Title:</b> " + titles[current_index]
        description_box.value = "<b>Description:</b> " + descriptions[current_index]
        keywords_box.value = "<b>Keywords:</b> " + keywords_list[current_index]
        
        # Update button states
        prev_button.disabled = (current_index == 0)
        next_button.disabled = (current_index == len(image_paths) - 1)
    
    # Button click event handlers
    def on_next_click(b):
        nonlocal current_index
        current_index += 1
        update_image()
    
    def on_prev_click(b):
        nonlocal current_index
        current_index -= 1
        update_image()
    
    # Link buttons to their respective click events
    next_button.on_click(on_next_click)
    prev_button.on_click(on_prev_click)
    
    # Display the first image and buttons
    update_image()
    
    # Arrange buttons and output
    button_box = HBox([prev_button, next_button], layout=widgets.Layout(justify_content='center'))
    
    # Display the image, title, description, and keywords in a clean, professional layout
    text_content = VBox([title_box, description_box, keywords_box], layout=widgets.Layout(align_items='flex-start'))
    image_and_description = HBox([output, text_content], layout=widgets.Layout(align_items='center', justify_content='center'))
    
    # Display everything
    display(VBox([image_and_description, button_box]))

```

```python
id = '6459efce0000000013010dc6'
rel_path = './../assets/downloaded_images/downloaded_images/'
image_paths = imagereturn(id,rel_path)

titles = [get_title_by_id(postsdf,id) for i in range(len(image_paths))]
print(titles)
descriptions = [get_description_by_id(postsdf,id) for i in range(len(image_paths))]
keywords_list = [get_keywords_by_id(postsdf,id) for i in range(len(image_paths))]

# Call the image carousel function
image_carousel(image_paths, titles, descriptions, keywords_list)
```
